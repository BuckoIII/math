### Trigonometry Exercises

2.  Convert the following angles in degrees to radians:
a)  12ᵒ
b)  65ᵒ
c)  200ᵒ


```python
# facts

# 360 degrees = 2pi radians
# 180 degresss = pi radians
# 1 degree = pi radians / 180

#  1 radian = 360 / 2pi
```

```python
answers = {}

answers['a'] = (m.pi / 180) * 12
answers['b'] = (m.pi / 180) * 65
answers['c'] = (m.pi / 180) * 200

def test_q2(answers):
    case = [12,65,200]
    result = np.deg2rad(case)
    format_ans = np.fromiter(answers.values(), dtype='float').round(3)

    if all(format_ans == np.array(result).round(3)):
        print('test passed')

    else:
        print('incorrect answers:', answers)
        print('correct answers:', result)

test_q2(answers)
```

```python
answers = {}

# cah - cos(x) = adj / hyp
# or hyp = adj / cos62
rad_p = 62*(m.pi/180)
answers['pq'] = 11 / np.cos(rad_p)

# can be solved with either pythagoras or tangent formula
answers['qr'] = (answers['pq']**2 - 11**2 )**(1/2)
answers['qr'] = np.tan(rad_p)  * 11


# Q = 180 - 90 - 62 = 28
answers['q'] = 28

print(answers)
```
